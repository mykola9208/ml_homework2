```python
import numpy as np
import random
```

<center>1. Знайти в датасеті таргет та видалити цю колонку з датасету (видаляти за індексом)</center>

```python
ds = np.genfromtxt('iris.data', delimiter=',')
```

```python
ds = np.delete(ds, 4, 1)
```

<center>2. Перетворити колонки, що залишились в 2D масив (або впевнитись, що це уже 2D масив)</center>

```python
ds = np.array(ds, ndmin = 2)
print('Num Dimension:', ds.ndim)
```

<center>3. Порахувати mean, median, standard deviation для 1-ї колонки</center>

```python
print('mean of the first column is ', np.mean(ds[:, 0]))
print('median of the first column is ', np.median(ds[:, 0]))
print('standard deviation of the first column is ', np.std(ds[:, 0]))
```

<center>4. Вставити 20 значень np.nan на випадкові позиції в масиві (при використанні звичайного рандому можуть накластись позиції, тому знайти рішення, яке гарантує 20 унікальних позицій)</center>

```python
ds = ds.flatten()
i=len(np.where(np.isnan(ds) == True))
while i<20:
    n = random.randint(0, ds.shape[0]-1)
    if np.isnan(ds[n]):
        pass
    else:
        ds[n] = np.nan
        i += 1
ds = np.reshape(ds, (150, 4))
ds
```

<center>5. Знайти позиції вставлених значень np.nan в 1-й колонці</center>

```python
np.where(np.isnan(ds[:, 0]) == True)
ds
```

<center>6. Відфільтрувати массив за умовою: значення в 3-й колонці > 1.5 та значения в 1-й колонці < 5.0</center>

```python
ds = ds[(ds[:, 2]> 1.5) & (ds[:, 0]< 5.0)]
```

<center>7. Замінити всі значення np.nan на 0</center>

```python
ds[np.isnan(ds)] = 0
```

<center>8. Порахувати всі унікальні значення в массиві та вивести їх разом із кількісю</center>

```python
unique, counts = np.unique(ds, return_counts = True)
for i in range (0, len(unique)):
    print(f'кількість значень {unique[i]} = {counts[i]}')
```

<center>9. Розбити масив по горизонталі на 2 рівні частини (не використовувати абсолютні числа, мають бути два массиви по 4 колонки)</center>

```python
if ds.shape[0] % 2 == 0:
    ds_new1, ds_new2 = np.split(ds, 2, axis = 0)
else:
    ds_new1, ds_new2 = np.split(ds[1:], 2, axis = 0)
    ds_new1 = np.vstack([ds[:1], ds_new1])
```

<center>10. Відсортувати обидва массиви по 1-й колонці: 1-й за збільшенням, 2-й за зменшенням</center>

```python
ds_new1 = ds_new1[ds_new1[:, 0].argsort()]
ds_new2 = ds_new2[ds_new2[:, 0].argsort()[:: -1]]
```

<center>11. Зібрати обидва массиви в одне ціле</center>

```python
ds = np.vstack([ds_new1, ds_new2])
```

<center>12. Знайти найбільш часто повторюване значення в массиві</center>

```python
unique[np.argmax(counts)]
```

<center>13. Написати функцію, яка б множила всі значення в колонці, які менше середнього значения в цій колонці, на 2, і ділила інші значення на 4.</center>

```python
def foo(arr, numb_col):
    mean = np.mean(arr[:, numb_col-1])
    arr[:, numb_col-1][(arr[:, numb_col-1] < mean)] *= 8
    arr[:, numb_col-1] /= 4
    
```
